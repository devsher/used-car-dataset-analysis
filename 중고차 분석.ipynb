```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',None)
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('car data.csv')
```

```python
df.head()
```

pd.set_option('display.max_columns', None) 코드는 pandas의 옵션을 설정하여 모든 열을 출력할 때 생략하지 않고 모두 표시하도록 합니다. 이는 데이터프레임의 모든 열을 볼 수 있게 해줍니다.

print(df)는 데이터프레임 df를 출력하는 명령입니다. 이를 통해 데이터프레임의 모든 행과 열이 표시됩니다.






```python
pd.set_option('display.max_columns', None)
print(df)
```

```python
df['Car_age'] = pd.Timestamp.now().year - df['Year']
df.head()
```

df.describe(percentiles=[0.25, 0.5, 0.75, 0.9, 0.95, 0.99]) 코드는 데이터프레임 df의 주요 통계 정보를 요약하여 보여줍니다. 추가로 지정한 백분위수에 해당하는 값들도 함께 표시됩니다.

이를 통해 데이터의 개수, 평균, 표준 편차, 최솟값, 1사분위수, 중앙값, 3사분위수, 90백분위수, 95백분위수, 99백분위수, 최댓값 등의 정보를 확인할 수 있습니다. 이를 통해 데이터의 분포와 중심 경향성, 이상치 여부 등을 파악할 수 있습니다.

```python
df.describe(percentiles=[0.25,0.5,0.75,0.9,0.95,0.99])
```


print(df.Fuel_Type.value_counts())는 데이터프레임 df의 'Fuel_Type' 열에 대해 각각의 값들의 빈도수를 출력합니다. 각 값이 몇 번 나타났는지를 세어서 보여줍니다.

print(df.Selling_type.value_counts())는 데이터프레임 df의 'Selling_type' 열에 대해 각각의 값들의 빈도수를 출력합니다. 역시 각 값이 몇 번 나타났는지를 세어서 보여줍니다.

print(df.Transmission.value_counts())는 데이터프레임 df의 'Transmission' 열에 대해 각각의 값들의 빈도수를 출력합니다. 마찬가지로 각 값이 몇 번 나타났는지를 세어서 보여줍니다.

이를 통해 각 열에 어떤 값들이 있는지와 각 값들이 데이터에서 얼마나 자주 등장하는지를 확인할 수 있습니다.

```python
print(df.Fuel_Type.value_counts())
print(df.Selling_type.value_counts())
print(df.Transmission.value_counts())
```

```python
#자동차 이름과 수량을 히스토그램 그래프로 보여 주기
```

```python
#자동차 이름과 수량을 히스토그램 그래프로 보여 주기
car_name_counts = df['Car_Name'].value_counts()

# 히스토그램 그리기
plt.figure(figsize=(20, 13))
car_name_counts.plot(kind='bar')
plt.xlabel('Car Name')
plt.ylabel('Count')
plt.title('car name and quantity')
plt.show()
```

```python
plt.figure(figsize=(17,12))
plt.subplot(2,2,1)
sns.barplot(df,x='Selling_type',y='Selling_Price')
plt.title('Selling Price Vs Seller Type')

plt.subplot(2,2,2)
sns.barplot(df,x='Transmission',y='Selling_Price')
plt.title('Selling Price Vs Transmission')

plt.subplot(2,2,3)
sns.barplot(df,x='Fuel_Type',y='Selling_Price')
plt.title('Selling Price Vs Fuel Type')
plt.figure(figsize=(8, 8))
```

```python
fuel_counts = df['Fuel_Type'].value_counts()

# 파이 그래프 그리기
plt.figure(figsize=(8, 8))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%')
plt.title('Fuel Type Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(data=df['Selling_Price'])
plt.title('Selling Price Distribution')

plt.xlabel('Selling Price')
plt.ylabel('')

plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.regplot(df, x='Car_age', y='Selling_Price')
plt.title('Selling Price Distribution by Car Age')

plt.xlabel('Car Age')
plt.ylabel('Selling Price')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

# 히스토그램 그리기
plt.hist(df['Year'], bins=10, edgecolor='black')

# 축 레이블 설정
plt.xlabel('Year')
plt.ylabel('Frequency')

# 그래프 제목 설정
plt.title('Car Year')

# 그래프 출력
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, x='Year', y='Selling_Price')
plt.title('Distribution of Selling Price by Year')
plt.xlabel('Year')
plt.ylabel('Selling Price')

plt.show()
```


위에서 분석한 데이터를 바탕으로 다음과 같은 결론을 도출할 수 있습니다:

>가격 예측 모델: 주어진 데이터셋을 사용하여 중고차의 가격을 예측할 수 있는 모델을 개발할 수 있습니다. 주행거리, 연료 종류, 차량 연도 등의 변수가 중고차 가격에 영향을 미칠 수 있으며, 이러한 변수를 고려하여 가격 예측 모델을 구축할 수 있습니다.


>연료 종류 분포: 데이터셋에는 주로 Petrol(휘발유)과 Diesel(디젤) 연료를 사용하는 차량이 포함되어 있습니다. CNG(압축 자연 가스) 연료를 사용하는 차량은 상대적으로 적은 수의 데이터만 포함되어 있습니다.


>판매 유형 분포: 데이터셋의 대부분 차량은 Individual(개인 판매)에 의해 판매되었습니다. Dealer(딜러)와 Trustmark Dealer(신뢰 딜러)에 의한 판매는 상대적으로 적은 수의 데이터를 가지고 있습니다.


>변속기 종류 분포: 주로 Manual(수동) 변속기를 가진 차량이 데이터셋에 포함되어 있으며, Automatic(자동) 변속기를 가진 차량도 일부 포함되어 있습니다.

>차량 연도 분포: 데이터셋에는 2003년부터 2018년까지 다양한 연식의 차량이 포함되어 있습니다. 가장 많은 차량은 2015년, 2014년, 2016년 등의 연식을 가지고 있습니다.

>차량 연식과 가격 관계: 분석 결과, 차량 연식이 높을수록 중고차 가격이 낮아지는 경향이 있습니다. 이는 차량의 노후화와 기술적인 발전에 따른 영향으로 해석될 수 있습니다.

따라서 이 데이터셋을 활용하여 중고차 가격 예측 모델을 개발하거나 중고차 시장의 트렌드와 특징을 파악할 수 있습니다. 또한 연료 종류, 판매 유형, 변속기 종류, 차량 연도 등의 변수를 고려하여 중고차 가격에 영향을 미치는 요인을 분석하고, 이를 차량 구매자나 판매자에게 유용한 정보로 활용할 수 있습니다.
